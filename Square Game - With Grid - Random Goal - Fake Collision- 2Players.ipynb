```python
import time
import pygame, sys
from pygame.locals import *
import random

pygame.init()
win_size = (600, 600)
win = pygame.display.set_mode(win_size)
pygame.display.set_caption("Square Game")
clock = pygame.time.Clock()

char_colour = [pygame.image.load('SquareChar1.png'), pygame.image.load('SquareChar2.png'), pygame.image.load('SquareChar3.png'), pygame.image.load('SquareChar4.png'), pygame.image.load('SquareChar5.png')]
playerone_win_message = pygame.image.load("PlayerOneWins.png")
playertwo_win_message = pygame.image.load("PlayerTwoWins.png")

line_coordinates = [100, 200, 300, 400, 500]
square_coordinates = [89, 189, 289, 389, 489]

WHITE = (255,255,255)
BLACK = (0,0,0)
RED = (255,0,0)
PURPLE = (252,194,252)

a = random.randint(0,4)
b = random.randint(0,4)
c = random.randint(0,4)

x = 89
y = 89

d = 489
e = 489

vel = 100

player1_colour_count = 0
player2_colour_count = 0


def redrawGameWindow():
    global player1_colour_count
    global player2_colour_count
    
    win.fill((0,0,0))
    
    for i in line_coordinates:
        pygame.draw.line(win, WHITE, (i, 100), (i, 500), 3)
        pygame.draw.line(win, WHITE, (100, i), (500, i), 3)
    
    
    if player1_colour_count + 1 >= 6:
        player1_colour_count = 0
    
    if player2_colour_count + 1 >= 6:
        player2_colour_count = 0
       
    win.blit(char_colour[a], (square_coordinates[b], (square_coordinates[c])))
    win.blit(char_colour[player1_colour_count], (x,y))
    win.blit(char_colour[player2_colour_count], (d,e))
    
    if square_coordinates[b] == x and square_coordinates[c] == y and char_colour[a] == char_colour[player1_colour_count]:
        win.blit(playerone_win_message, (140,25))
    
    if square_coordinates[b] == d and square_coordinates[c] == e and char_colour[a] == char_colour[player2_colour_count]:
        win.blit(playertwo_win_message, (140,510))
    
    pygame.display.update()

    
run = True
while run:
    clock.tick(30)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
  
    keys = pygame.key.get_pressed()

    if keys[pygame.K_LEFT] and x > vel:
        x -= vel
        player1_colour_count += 1
        pygame.time.wait(100)
    
    if keys[pygame.K_RIGHT] and x < 450:
        x += vel
        player1_colour_count += 1
        pygame.time.wait(100)
        
    if keys[pygame.K_UP] and y > vel:
        y -= vel
        player1_colour_count += 1
        pygame.time.wait(100)
    
    if keys[pygame.K_DOWN] and y < 450:
        y += vel
        player1_colour_count += 1
        pygame.time.wait(100)
        
    if keys[pygame.K_a] and d > vel:
        d -= vel
        player2_colour_count += 1
        pygame.time.wait(100)
    
    if keys[pygame.K_d] and d < 450:
        d += vel
        player2_colour_count += 1
        pygame.time.wait(100)
        
    if keys[pygame.K_w] and e > vel:
        e -= vel
        player2_colour_count += 1
        pygame.time.wait(100)
    
    if keys[pygame.K_s] and e < 450:
        e += vel
        player2_colour_count += 1
        pygame.time.wait(100)
        
    redrawGameWindow()

pygame.quit()
```
